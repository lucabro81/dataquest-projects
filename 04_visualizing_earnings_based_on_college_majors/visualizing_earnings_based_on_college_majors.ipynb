## Introduction

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- `Rank` - Rank by median earnings (the dataset is ordered by this column).
- `Major_code` - Major code.
- `Major` - Major description.
- `Major_category` - Category of major.
- `Total` - Total number of people with major.
- `Sample_size` - Sample size (unweighted) of full-time.
- `Men` - Male graduates.
- `Women` - Female graduates.
- `ShareWomen` - Women as share of total.
- `Employed` - Number employed.
- `Median` - Median salary of full-time, year-round workers.
- `Low_wage_jobs` - Number in low-wage service jobs.
- `Full_time` - Number employed 35 hours or more.
- `Part_time` - Number employed less than 35 hours.

#### What percent of majors are predominantly male? Predominantly female?

```python
major_women = recent_grads[recent_grads['ShareWomen'] > 0.5][['Major', 'Total', 'Men', 'Women']]
major_women.sort_values('Women', ascending=False)
```

```python
print('Percent of majors predominantly female:', (len(major_women)/len(recent_grads))*100)
```

```python
major_men = recent_grads[recent_grads['ShareWomen'] < 0.5][['Major', 'Total', 'Men', 'Women']]
major_men.sort_values('Men', ascending=False)
```

```python
print('Percent of majors predominantly male:', (len(major_men)/len(recent_grads))*100)
```

#### What's the most common median salary range?

```python
import math
import operator

max_median = recent_grads['Median'].max()
min_median = recent_grads['Median'].min()
#interval = (max_median - min_median) / 25;
interval = 1000
interval_dict = {}

for val in recent_grads['Median']:
    val = math.floor(val / interval)
    if val in interval_dict:
        interval_dict[val] += 1;
    else:
        interval_dict[val] = 1;
    
sorted_median = sorted(interval_dict.items(), key=operator.itemgetter(1), reverse=True)
lower_edge = (sorted_median[0][0]*interval)
upper_edge = ((sorted_median[0][0]+1)*(interval))
print('Most common median salary range', lower_edge, '-', upper_edge);
```
